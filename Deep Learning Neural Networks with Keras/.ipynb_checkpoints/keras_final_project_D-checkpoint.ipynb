```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn import model_selection
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Load the dataset
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
concrete_data_columns = concrete_data.columns
```

```python
# Dived the data into predictors and target info
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
# Normalize the predictors
predictors_norm = (predictors - predictors.mean()) / predictors.std()
```

```python
# Train test split
X_train, X_test, y_train, y_test = model_selection.train_test_split(predictors_norm, target, test_size=0.3, random_state=22)
print('X_train {0}, y_train {1}'.format(X_train.shape, y_train.shape))
print('X_test {0}, y_test {1}'.format(X_test.shape, y_test.shape))
```

```python
n_cols = predictors_norm.shape[1] # number of predictors
```

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu', input_shape=(10,)))
    model.add(Dense(10, activation='relu', input_shape=(10,)))
    model.add(Dense(1))
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model()
```

```python
model.fit(predictors_norm, target, validation_split=0.3, epochs=50, verbose=2)
```

```python
# Make predictions from the test set
preds = model.predict(X_test)
```

```python
# Mean squared error of the y truth compared to the predicted output
mean_squared_error(y_test, preds)
```

```python
# Mean squared error list
mse_list = []
for i in range(50):
    model.fit(predictors_norm, target, validation_split=0.3, epochs=50, verbose=2)
    preds = model.predict(X_test)
    mse_list.append(mean_squared_error(y_test, preds))
```

```python
# Mean of the MSE
np.mean(mse_list)
```
